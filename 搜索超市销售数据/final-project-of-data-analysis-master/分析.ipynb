```python
import pandas as pd
import matplotlib as plt
```

```python
data=pd.read_csv("order-14.3.csv",encoding='gbk')
display(data)
```

```python
# 畅销类别
data.groupby("类别ID")["销量"].sum().reset_index().sort_values(by="销量",ascending=False).head(10)
```

```python
# 畅销商品
pd.pivot_table(data,index="商品ID",values="销量",aggfunc="sum").reset_index().sort_values(by="销量",ascending=False).head(10)
```

```python
# 计算销售额
data["销售额"] = data["单价"]*data["销量"]
data
```

```python
# 各个门店销售额分别相加
data.groupby("门店编号")["销售额"].sum()
```

```python
# 算销售额占比
销售额占比 = data.groupby("门店编号")[["销售额"]].sum()/data["销售额"].sum()
销售额占比
```

```python
# 替换列名
销售额占比.columns = ['销售额占比']
销售额占比
```

```python
# 制作饼图
# 设置figure_size尺寸
plt.rcParams['figure.figsize'] = (16.0, 8.0)
# 用来设置字体样式以正常显示中文标签
plt.rcParams['font.sans-serif']=['SimHei']
# 默认是使用Unicode负号，设置正常显示字符，如正常显示负号
plt.rcParams['axes.unicode_minus']=False    
plt.rcParams['font.size'] = 15

(data.groupby("门店编号")["销售额"].sum()/data["销售额"].sum()).plot.pie()
```
