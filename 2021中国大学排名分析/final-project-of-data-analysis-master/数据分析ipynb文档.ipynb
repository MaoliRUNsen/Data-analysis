## 各省市大学数量

```python
## 统计各省市大学数量
g = df.groupby('省市')    
df0 = g.count()['排名']  #统计数量
df_number = df0.copy()
df_number.sort_values(ascending=True, inplace=True)
df_number
```

各省市大学数量地图分布 
[参考](https://www.freesion.com/article/7430533154/)

```python
#各省市大学数量地图分布 
cityname = df0.index.tolist() #收集省市名
number = df0.values.tolist() #各省市大学数量
map = (
        Map()
        .add('', [list(i) for i in zip(cityname, number)], 'china')
        .set_global_opts(
            title_opts=opts.TitleOpts(title='中国各省市大学数量分布'),
            visualmap_opts=opts.VisualMapOpts(max_=40, split_number=8, is_piecewise=True), #根据最大数量的省市（江苏：38）设置最大值，并设置一个合适的间隔数
        )
    )
map.render_notebook()
```

## 各省市大学平均分

```python
## 统计各省市平均分
df1 = g.mean()['总分']  #统计平均分
df1 = df1.round(1)      #保留一位小数
df1
```

```python
df3 = pd.concat([df0, df1], join='outer', axis=1)   #合并数量与平均分
df3.columns = ['数量', '平均分']                    #添加列名
df3.sort_values(by=['平均分'], ascending=False, inplace=True)  #按平均分倒叙排序
df3
```

各省市大学平均分柱状图

```python
省市=df3.index.tolist()
平均分=df3['平均分'].values.tolist()
bar = (
    Bar()
    .add_xaxis(省市)
    .add_yaxis("平均分", 平均分)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="各省市大学平均分"),
        yaxis_opts=opts.AxisOpts(name="省"),
        xaxis_opts=opts.AxisOpts(name="分"),)
    .reversal_axis()
    .set_series_opts(label_opts=opts.LabelOpts(position="right")))
    
bar.render_notebook()
```

## flask搭建框架

```python
#院校类型
类型名称 = list(df.类型.dropna().unique())
类型名称
```

```python
def i(the_type):
    dfs = df.query("类型=='{}'".format(the_type))
    df_summary = dfs.groupby("省市").agg({"学校名称":"count","总分":"mean"}).sort_values(by = "总分",ascending = False )
    df_summary.columns=['学校名称','平均分']
    return df_summary 
i('理工')
```

```python
#可视化
def i(the_type):
    dfs = df.query("类型=='{}'".format(the_type))
    df_summary = dfs.groupby("省市").agg({"学校名称":"count","总分":"mean"}).sort_values(by = "总分",ascending = False )
    df_summary.columns=['学校名称','平均分']
    fig = dfs.iplot(kind="bar", x="省市", y="总分", asFigure=True)
    return fig
i('理工') 
```
