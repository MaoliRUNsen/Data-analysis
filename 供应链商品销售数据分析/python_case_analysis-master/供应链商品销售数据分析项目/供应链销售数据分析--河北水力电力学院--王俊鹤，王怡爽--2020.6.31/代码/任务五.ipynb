# 一月份分析

```python
import pandas as pd
```

```python
data=pd.read_excel('./data/供应链商品销售数据.xlsx')
```

```python
data.columns
```

```python
data1=data.loc[data['销售点类型']=='社区店']
```

```python
data1.loc[data1['销售月份']=='January']
```

```python
data2=data1.loc[data1['销售月份']=='January']
```

```python
data2.groupby(['商品代号','销售月份','销售点类型']).sum()
```

```python
data3=data2[['商品代号','销售额（万元）']].groupby(by='商品代号')
```

```python
data3.sum()
```

```python
data4=data3.sum()
```

```python
data4.index
```

```python
from pyecharts.charts import Pie
import pyecharts.options as opts

num=data4['销售额（万元）']
lab=data4.index
(
      Pie(init_opts=opts.InitOpts(width='720px',height='320px'))#画布大小
    .add(series_name='',data_pair=[(j,i)for i,j in zip(num,lab)])
).render_notebook()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']= 'SimHei'  #中文显示
plt.rcParams['axes.unicode_minus']=False

plt.figure(figsize=(15,15))
plt.style.use('ggplot')
plt.pie(data4['销售额（万元）'],labels=data4.index,autopct='%.2f %%', wedgeprops={'width':0.6,'edgecolor':'w'})

plt.title('一月份各销售占比')

plt.show()
```
