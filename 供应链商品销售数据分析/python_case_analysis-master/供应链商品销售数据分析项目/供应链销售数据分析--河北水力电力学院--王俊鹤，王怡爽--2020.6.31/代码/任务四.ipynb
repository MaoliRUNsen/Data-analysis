```python
import numpy as np
import pandas as pd
tmp = pd.read_excel('./data/供应链商品销售数据.xlsx') #读取excel表格

# 任务4 ：分析销售季节性特点，据以安排店员年休假或培训时间，以及订货量控制

tmp2_gb = pd.pivot_table(tmp[['销售月份', '销售点类型', '销售额（万元）']], 
               index=['销售点类型', '销售月份'],aggfunc=np.sum)  # 创建透视表
tmp2_gb2 = tmp2_gb.loc['社区店']  # 社区店销售月份与销售额关系
ind = ['January','February','March','April','May','June','July','August','September','October','November','December']
number2 = tmp2_gb2.loc[ind]     # 社区店按月份排序

from matplotlib import pyplot as plt

#设置图形大小

plt.figure(figsize=(15,12),dpi = 80)
# 设置matplotlib正常显示中文和负号
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
#数据
a = ['January','February','March','April','May','June','July','August','September','October','November','December']
b = number2['销售额（万元）']

height = 0.2
a1 = list(range(len(a)))
a2 = [i+height for i in a1]#坐标轴偏移
#绘图
plt.barh(range(len(a)),b,height= height,label = "社区店",color = "#FFC125")

#绘制网格
plt.grid(alpha = 0.4)

#y轴坐标刻度标识
plt.yticks(a2,a,fontsize = 14)

#添加图例
plt.legend()

#添加横纵坐标，标题
plt.xlabel("销售额（万元）",fontsize = 16)
plt.ylabel("销售月份",fontsize = 16)
plt.title("社区店季度与各月份销售额展示",fontsize = 24)

#显示图形
plt.show()
```

```python
# 绘制饼图
from pyecharts.charts import Pie
import pyecharts.options as opts

num = [1140.0, 1216.0, 1265.0, 1360.5]
lab = ['第一季度', '第二季度', '第三季度', '第四季度']

(
     Pie(init_opts=opts.InitOpts(width='720px', height='320px'))
    .add(series_name='', data_pair=[(j, i) for i, j in zip(num, lab)],radius=['40%', '80%'])   # radius为设置空心圆 
    .set_global_opts(title_opts=opts.TitleOpts(title='CBD店季度销售情况')) 
).render_notebook()
```
